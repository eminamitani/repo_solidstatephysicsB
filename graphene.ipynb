```python
import plotly.graph_objs as go
import numpy as np

import math

# バンド分散のデータを生成
x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)
X, Y = np.meshgrid(x, y)

def H(t,a,kx,ky):
    f=-t*(np.exp(1.0j*kx*a)+2.0*np.exp(-1.0j*kx*a*0.5)*math.cos(math.sqrt(3)*ky*a*0.5))
    ham=[[0.0, f],[np.conjugate(f), 0.0]]
    return ham

eigenc=np.zeros((100,100))
eigenv=np.zeros((100,100))

for ikx,kx in enumerate(x):
    for iky,ky in enumerate(y):
        ham=H(1.0,1.0,kx,ky)
        eig, l=np.linalg.eig(ham)
        
        eigenc[ikx,iky]=eig[0].real
        eigenv[ikx,iky]=eig[1].real


# Surfaceオブジェクトを作成
#surface_upper = go.Surface(x=X, y=Y, z=eigenc)
#surface_lower=  go.Surface(x=X, y=Y, z=eigenv,colorbar=None)

# レイアウトを作成
layout = go.Layout(
    title='3D Band Structure',
    scene=dict(
        xaxis=dict(title='kx'),
        yaxis=dict(title='ky'),
        zaxis=dict(title='Energy')
    )
)

# Figureオブジェクトを作成
#fig = go.Figure(data=[surface_upper,surface_lower], layout=layout)
fig = go.Figure()

# Surface1の追加
fig.add_trace(go.Surface(x=X, y=Y, z=eigenc, colorscale='Viridis', name='Surface1',cmax=3,cmid=0))

# Surface2の追加
fig.add_trace(go.Surface(x=X, y=Y, z=eigenv, colorscale='Viridis', name='Surface2',cmax=3,cmid=0))

# グラフを表示
fig.show()
```
