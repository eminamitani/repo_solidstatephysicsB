```python
import numpy as np
import matplotlib.pyplot as plt

#3つの波の合成を定義
def func(t,f0,f1,f2):
    return 1.0*np.sin(2.0*np.pi*t*f0) + 0.5*np.cos(2.0*np.pi*t*f1)+0.3*np.cos(2.0*np.pi*t*f2)
```

```python
# サンプリングレート
N = 20000
ts = 0.0
te = 1.0
f0=100
f1=90
f2=50
t = np.linspace(ts, te, N)
#f0,f1,f2は音波の場合のHz単位の周波数に相当
#3つの成分が混じった音
x = func(t,f0,f1,f2)

# 離散フーリエ変換
X = np.fft.fft(x)
dt = t[1] - t[0] # 時間刻み
f = np.fft.fftfreq(N, dt) 

# データをプロット
fig, (ax01, ax02) = plt.subplots(nrows=2, figsize=(6, 8))
plt.subplots_adjust(wspace=0.0, hspace=0.6)

#見やすいように0.2まで
ax01.set_xlim(0.0,0.2)
ax01.set_xlabel('t')
ax01.set_ylabel('x')
ax01.plot(t, x) # 入力信号

ax02.set_xlim(-120, 120)
ax02.set_xlabel('frequency (Hz)')
ax02.set_ylabel('|X|/N')
ax02.plot(f, np.abs(X)/N) # 振幅スペクトル

plt.savefig('Fourier.png',dpi=600)
```
